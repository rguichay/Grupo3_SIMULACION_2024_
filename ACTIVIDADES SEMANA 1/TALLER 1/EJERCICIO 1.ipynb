# Dada la siguiente función f(x) = 10e^(x/2) Cos(2x) en el intervalo [0,10].
# Graficar la función y encontrar las raices.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Definir la función
def f(x):
    return 10 * np.exp(x / 2) * np.cos(2 * x)

# Crear un rango de valores de x entre 0 y 10
x = np.linspace(0, 10, 1000)

# Calcular los valores de f(x) para estos valores de x
y = f(x)

# Graficar la función
plt.plot(x, y, label='f(x) = 10 * exp(x / 2) * cos(2x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Gráfica de f(x) en el intervalo (0,10)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Encontrar las raíces por el método punto a punto
# Definir un rango de puntos iniciales para la búsqueda de raíces
initial_guesses = np.linspace(0, 10, 100)

# Encontrar las raíces
roots = []
for guess in initial_guesses:
    root = fsolve(f, guess)[0]
    if 0 < root < 10:  # Asegurarse de que la raíz esté en el intervalo
        roots.append(root)

# Eliminar duplicados
roots = np.unique(np.round(roots, decimals=4))

print("Las raíces de la función en el intervalo (0,10) son:", roots)
```
